```python
import pandas as pd
from sklearn.datasets import load_wine

wine = load_wine(as_frame=True)
df = wine.data
df['target'] = wine.target
df.head()
df.tail()
df = df.fillna(0)
```

```python
summary = df.groupby('target')['alcohol'].mean().reset_index()
display(summary)
```

```python
summary = summary.sort_values('alcohol', ascending=False)
display(summary)
```

```python
pivot = pd.pivot_table(df,
                       values=['alcohol', 'color_intensity'],
                       index='target',
                       aggfunc='mean')
display(pivot)
```

```python
summary.to_csv('wine_summary.csv', index=False)
```
